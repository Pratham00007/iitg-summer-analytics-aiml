```python
plt.plot(gas.Year,gas.USA)
plt.show()
```

```python
plt.plot(gas.Year,gas.USA)
plt.plot(gas.Year,gas.Canada)
plt.plot(gas['Year'],gas['South Korea'])

plt.show()
```

```python

plt.title('Gas Price over time (in USD)')
plt.plot(gas.Year,gas.USA)
plt.plot(gas.Year,gas.Canada)
plt.plot(gas['Year'],gas['South Korea'])

plt.legend()

plt.show()
```

```python

plt.title('Gas Price over time (in USD)')
plt.plot(gas.Year,gas.USA,label='United States')
plt.plot(gas.Year,gas.Canada)
plt.plot(gas['Year'],gas['South Korea'])

plt.legend()

plt.show()
```

```python

plt.title('Gas Price over time (in USD)')
plt.plot(gas.Year,gas.USA,label='United States')
plt.plot(gas.Year,gas.Canada)
plt.plot(gas['Year'],gas['South Korea'])

plt.xticks(gas.Year[::3])

plt.legend()

plt.show()
```

```python

plt.title('Gas Price over time (in USD)')
plt.plot(gas.Year,gas.USA,'b.-',label='United States')
plt.plot(gas.Year,gas.Canada,'r.-')
plt.plot(gas['Year'],gas['South Korea'],'g.-')
plt.plot(gas.Year,gas.Australia,'y.-')

plt.legend()

plt.xticks(gas.Year[::3])

plt.show()
```

```python

plt.title('Gas Price over time (in USD)')

# plt.plot(gas.Year,gas.USA,'b.-',label='United States')
# plt.plot(gas.Year,gas.Canada,'r.-')
# plt.plot(gas['Year'],gas['South Korea'],'g.-')
# plt.plot(gas.Year,gas.Australia,'y.-')


for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country],marker='.')
plt.legend()

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollar')

plt.show()
```

```python

plt.title('Gas Price over time (in USD)')

# plt.plot(gas.Year,gas.USA,'b.-',label='United States')
# plt.plot(gas.Year,gas.Canada,'r.-')
# plt.plot(gas['Year'],gas['South Korea'],'g.-')
# plt.plot(gas.Year,gas.Australia,'y.-')

country_to_look=['Australia','USA','Canada','South Korea']

for country in gas:
    if country in country_to_look:
        plt.plot(gas.Year,gas[country],marker='.')
plt.legend()

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollar')

plt.show()
```
